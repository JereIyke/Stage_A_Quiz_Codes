```python
#Question 1
#What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?
df_mean = df.Y2017.mean()
print(round(df_mean, 2))
df_std = df.Y2017.std()
print(round(df_std, 2))
```

```python
#Question 4
#Answer: By visual inspection, the answer is Y2014.
#Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?
df1 = df.groupby('Element')['Y2014', 'Y2015', 'Y2016','Y2017', 'Y2018'].sum()
print(df1)
```

```python
#Question 6
#Which year had the least correlation with ‘Element Code’?

#select the relevant features and the dependent variable from the dataset
df_select = df.iloc[:, 4:12]
print(df_select)
df_Remove = df_select.drop(columns = df_select.columns[[1, 2]])
print(df_Remove)

# Compute the correlation coefficients between each feature and the dependent variable
correlations = df_Remove.corr()['Element Code'].drop('Element Code')  # Drop the dependent variable itself

#Find the least correlation coefficient
least_corr_feature = correlations.idxmin()
least_corr_coefficient = correlations.min()

print("Least Correlated Feature:", least_corr_feature)
print("Least Correlation Coefficient:", least_corr_coefficient)
```

```python
#Question 7
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

#Step1: Perform a groupby operation by 'Area' and calculate the total sum in each 'Area' in Y2017
df2 = df.groupby('Area')['Y2017', 'Area'].sum()
print(df2)

#Step2: Sort df2 by 'Y2017' and get the Area with the 7th lowest sum using the head() function 
#The answer is Guinea Bissau
sorted_df2 = df2.sort_values(by='Y2017', ascending=True)
print(sorted_df2.head(7))
```

```python
#Question 8
#What is the total number and percentage of missing data in 2014 to 3 decimal places?

#Get the total number of missing data in 2014 
df1 = df.Y2014.isnull().sum()
print(df1)

#Get the total number of non-missiong data
df2 = df.Y2014.notnull().sum()
print(df2)

#Calculate the percentage of missing data to 3 decimal places
df_percentage = (df1/df2) * 100
print(round(df_percentage, 3))
```
